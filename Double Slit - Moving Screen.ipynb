A verification of the claim made that in an electron double slit experiment with the screen movable in the direction parallel to the line joining the slits, the pattern obtained on the screen is similar to that expected classically.


The slit through which the electron passes can be determined by measuring the momentum of the screen. However, this precise measurement of momentum restricts the precision of the measurement of the center of the screen to d/2, where d is the distance between the second maximas of the diffraction pattern (Heisenberg's Uncertainty Principle). 


This uncertainty, in turn, makes the recorded position of the electron hit uncertain (since this position is measured with respect to the center of the screen). The uncertainty, when incorporated into the diffraction pattern, distorts it so that it becomes the same as the classical case.

```python
import numpy as np
import matplotlib.pyplot as plt

width = 4
partitions = 1000

x = np.linspace(-width, width, partitions)
px = np.sinc(x)

mx = 2.5

plt.figure(1)
plt.plot(x, px, 'r-')
plt.plot(mx, np.sinc(mx), 'bo')
plt.xlabel("Position")
plt.ylabel("Probability")
plt.show()
```

```python
del_d = mx
px_averaged_list = []

i = 0

for posOrigin in x:
    count = 0
    px_averaged_list.append(0)
    
    for pos2 in x:   
        if abs(posOrigin - pos2) < mx:
            px_averaged_list[i] += np.sinc(pos2) 
            count += 1
    px_averaged_list[i] /= count
    i += 1
    
plt.figure(2)
plt.plot(x, px_averaged_list, 'r--')
plt.xlabel("Position")
plt.ylabel("Probability")
plt.show()

```
